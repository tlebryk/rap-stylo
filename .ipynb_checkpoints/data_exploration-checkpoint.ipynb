```python
import json
import os
import pandas as pd
from nltk import FreqDist
from nltk.corpus import stopwords 
import re
# from nltk import 
from sklearn.feature_extraction.text import CountVectorizer as CV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
import numpy as np
```

```python
#input path, return list with of albums in json format
# each json entry is a song. 
def json_extract(path):
    data_list=[]
    for file in os.listdir(path): 
        if file[-5:] == '.json':
            with open(path+file, 'r') as f: 
                data = json.load(f)
                data_list.append(data)
    return data_list
```

```python
path = 'data/drake/'
drake=json_extract(path)
path='data/quentin_miller/'
quentin=json_extract(path)
```

```python
corpus = []
song_titles = []
iyrtitl= []
drake_count=0
for album in drake: 
    for song in album: 
        if len(song["lyrics"]) < 10: 
            continue
        # keep track of "If you're..." to put in test set (iytitl is subject to ambiguous authorship)
        if song["album"]=="If You’re Reading This It’s Too Late ":
            iyrtitl.append(song['title'])
        corpus.append(song['lyrics'])
        song_titles.append(song['title'])
        drake_count+=1
for album in quentin: 
    for song in album: 
        if len(song["lyrics"]) < 10: 
            continue
        corpus.append(song['lyrics'])
        song_titles.append(song['title'])
```

```python
#remove blatant identifiers that are unique to a rapper 
stops=['drizzy', 'drake', 'quentin', 'miller', 'ovo', 'champagne', 'papi','toronto', 'atlanta', 'october', '6', 'october']
```

```python
cv = CV(stop_words=stops, ngram_range=(1,2))
matrix=cv.fit_transform(corpus).toarray()
```

```python
qd_df= pd.DataFrame(matrix, index=song_titles, columns=cv.get_feature_names())
# code zero for drake, one for Quentin
_class = np.zeros(len(corpus))
_class[drake_count:] = 1
qd_df['_class'] = _class
qd_df.drop_duplicates(inplace=True)
# save iyrytitl songs to place in test set
df2=qd_df.loc[iyrtitl]
qd_df.drop(iyrtitl, axis=0, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(qd_df.drop("_class", axis=1), np.array(qd_df['_class']), test_size=.1, stratify=qd_df['_class'])
```

```python
X_test=X_test.append(df2.drop("_class", axis=1))
y_test=np.append(y_test, (np.zeros(len(df2))))
```

```python
#spase matrix so choose l1 considering many words likely don't contribute much to prediction
model = LogisticRegressionCV(cv=5, penalty='l1', solver='liblinear')
model.fit(X_train, y_train)
# print("Logistic regression's tra"model.score(X_train, y_train))
```

```python
cdf = pd.DataFrame(model.coef_.T, X_train.columns, columns=['Coefficients'])
```

```python
#show strongest predictors of drake
cdf.sort_values(['Coefficients'])[:10]
```

```python
#show strongest predictors of Quenton Miller
cdf.sort_values(['Coefficients'])[-10:]
```

```python
print("test set accuracy: {}".format(model.score(X_test, y_test)))
print("training accuracy: {}".format(model.score(X_train, y_train)))
```

```python
probs=pd.DataFrame(model.predict_proba(X_test), index=X_test.index, columns=["drake", "quentin"])
probs['prediction']=model.predict(X_test)
probs['true_vals']=y_test     
probs
```

```python
print("best average cv score: _C={} with : {}".format(model.C_[0] , np.max(model.scores_[1.0].sum(axis=0)/5) ))
```
